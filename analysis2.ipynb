# **Audio Analysis**

#### Bhajan Analysis using Audio Features

```python
# Bhajan Audio Analysis
audio_path1 = os.path.join(BASE_DIR, 'Music', 'bhajan','[SPOTDOWNLOADER.COM] Aarambh Hai Prachand.mp3')
audio_data1, sr1 = librosa.load(audio_path1)

# waveform
plt.figure(figsize=(24, 16))
plt.subplot(4, 2, 1)
librosa.display.waveshow(audio_data1, sr=sr1)
plt.title('Waveform')

# Spectrogram
D1 = librosa.amplitude_to_db(np.abs(librosa.stft(audio_data1)), ref=np.max)
plt.figure(figsize=(26, 16))
plt.subplot(4, 2, 1)
librosa.display.specshow(D1, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-frequency power spectrogram')

#
```

#### **Summary** 
- **Waveform** : At the beginning of the audio, the amplitude is low and then it increases. Then it holds the amplitude for most of the audio and it has some sudden drops in amplitude.
- **Spectrogram** : The spectrogram shows the frequency content of the audio. The audio has a lot of low-frequency content compare to high-frequency content throughout the audio. The low-frequency content is more dominant in the audio.

# Dataset Analysis

```python
data_path1= os.path.join(os.getcwd(), '..', 'automations', 'thumri-nazrul-rabindra-kawali_segment_more_features.csv')
data_path2= os.path.join(os.getcwd(), '..', 'automations', 'gacmis_song_segment_more_features.csv')
data_path1, data_path2
```

```python
df1 = pd.read_csv(data_path1)
df2 = pd.read_csv(data_path2)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df = df.dropna()
df 
```

```python
# remove the songs with songtype = "thumri"
df = df[df.songtype != "Garhwali"]
df = df[df.songtype != "sufi"]
df["songtype"].value_counts()
```

```python
# reduce the number of rows for each songtype to 1000
df = df.groupby('songtype').apply(lambda x: x.sample(2100)).reset_index(drop=True)
df["songtype"].value_counts()
```

```python
# one hot encoding
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoder.fit(df[['songtype']])
hotted_data = encoder.fit_transform(df[['songtype']]).toarray()
hotted_labels = encoder.categories_[0]

hotted_data_df = pd.DataFrame(hotted_data, columns=hotted_labels)
hotted_data_df
```

```python
updated_df = pd.concat([df, hotted_data_df], axis=1)
updated_df
```

```python
# heatmap
plt.figure(figsize=(16, 16))
sns.heatmap(updated_df.drop(columns=['songtype','audio_name']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
